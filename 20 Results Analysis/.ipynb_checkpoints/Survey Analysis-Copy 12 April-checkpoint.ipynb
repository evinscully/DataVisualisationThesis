### 1.A. Importing and Tidying Data

First the survey responses and true values (tv) are imported from .csv files and saved as dataframes. Then the "Block" column is made and labelled appropriately.

The true values are then merged with the appropriate dataframe so as to be easily comparable.

```python
df_tv = pd.read_csv('TrueValues.csv')
df_b1 = pd.read_csv('Block 1 Responses.csv')
df_b2 = pd.read_csv('Block 2 Responses.csv')
df_b3 = pd.read_csv('Block 3 Responses.csv')
df_b4 = pd.read_csv('Block 4 Responses.csv')

df_b1['Block']=1
df_b2['Block']=2
df_b3['Block']=3
df_b4['Block']=4

df_b1 = pd.merge(df_b1, df_tv, how='left', left_on='Block', right_on='block')
df_b2 = pd.merge(df_b2, df_tv, how='left', left_on='Block', right_on='block')
df_b3 = pd.merge(df_b3, df_tv, how='left', left_on='Block', right_on='block')
df_b4 = pd.merge(df_b4, df_tv, how='left', left_on='Block', right_on='block')
```

Concatenate all the datasets together

```python
df = pd.concat([df_b1, df_b2, df_b3, df_b4])
df = df.reset_index(drop=True)
```

Remove the example questions

```python
df.drop(['exp1','exp2','exp3','exp4'], axis=1, inplace=True)
```

Convert two angle questions into percentages from degrees

```python
df['angle1'] = df['angle1']/3.6
df['angle2'] = df['angle2']/3.6
```

Convert the datetime column from a string into a date

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

View a list of all the column names

```python
list(df.columns)
```

Create a Total Error field, which is the cumulative sum of the absolute error per participant.

```python
# They are first set to 0 so that if the cell is reran it doesn't sum it again
df['Total_Error'] = 0


# iloc 13-25 refers to columns number 13-25 which are the guess columns. Then subtracted from the true values, which are 16 columns later
for i in range(13,25):
    df['Total_Error'] += abs(df.iloc[:, i] - df.iloc[:, i+16])
    print(df.iloc[0, i], df.iloc[0, i+16]) # Printing one set of true values and judgements for sanity check
```

```python
# To check  that the iloc columns are correct they are printed here:
print(df.iloc[0,13:25])
print(df.iloc[0,13+16:25+16])
```

Tidying some categorical outliers

```python
df['education'].replace(['GCSE ?', 'ACA '], np.NaN, inplace=True)
df['education'] = df['education'].astype(pd.api.types.CategoricalDtype(categories = 
                                                                       ["Leaving Certificate", "Ordinary Bachelor Degree", "Honours Bachelor Degree", "Masters Degree", "Doctoral Degree", ""]
                                                                      ,ordered=True))

df['gender'].replace(['Non binary ', 'None'], 'Other', inplace=True)
df['eyecolour'].replace(['Bluey greeny yellowy grey', 'Greyish blue ', 'Hazel', 'grey',
 'White around the edges, black in the middle', 'blue-green', 'hazel'], 'Other', inplace=True)
set(df.gender)
set(df.eyecolour)
```

Trimming Data

```python
print('This is the amount of observations before trimming:',len(df))

high = 0.90
low = 0.1

upperq = df['Total_Error'].quantile(high)
lowerq = df['Total_Error'].quantile(low)

# tdf = Trimmed Data Frame
df['Trimmed'] = df['Total_Error'].between(lowerq,upperq)
tdf = df[df['Trimmed']]

trimDictionary = {True: 'Kept', False: 'Trimmed'}
df['Trimmed'] = df['Trimmed'].replace(trimDictionary)

print(upperq)
print(lowerq)


print('This is the amount of observations after trimming:',len(tdf))
```

```python
g=sns.swarmplot( x=[""]*len(df), y=df['Total_Error'], hue=df['Trimmed'],size=8, palette=[pal[0],pal[2]])
sns.despine()
g.set(ylabel='Total Error per Participant')

plt.savefig('../output/03Btrim.png', dpi = 300)
```

### 1.B. Reshaping Data

The dataset will now be unpivoted into a "long" format. This means that each observation is not on a per user basis, but on a *per answer* basis. This is necessary to make have the data sub divided into smaller slices and make comparisons between each chart.

```python
df_long = pd.melt(tdf, id_vars=['timestamp', 'Block', 'age', 'gender', 'education', 'device', 'eyecolour', 'colourblind', 'learntype', 'vistools'],\
                   value_vars=['pie', 'donut', 'treemap', 'bar', 'stacked', 'funnel', 'angle1', 'angle2', 'arc1', 'arc2', 'area1', 'area2',],\
                   var_name='charttype', value_name='judgement')

#Create a label called "chart_id" to merge when combined with true values later
df_long['chart_id'] = df_long['charttype'] + df_long['Block'].apply(str)

#Combine the angle1 & angle 2 into one angle chart type. Do the same for arc and angle.
df_long['charttype'].replace(to_replace=['angle1', 'angle2', 'arc1', 'arc2', 'area1', 'area2'],
                             value=['angle','angle','arc','arc','area','area'],
                             method='string', inplace=True)

print('Here is a list of the chartypes :',set(df_long.charttype))
```

Create longform of true values (tv) and merge it with long form of the dataframe

```python
df_tv_long = pd.melt(df_tv, id_vars=['block'],
                     value_vars=['pie_tv','donut_tv','treemap_tv','bar_tv','stacked_tv','funnel_tv','angle1_tv','angle2_tv','arc1_tv','arc2_tv','area1_tv','area2_tv'],
                     var_name='tvcharttype', value_name='trueValue')

df_tv_long['tvcharttype'] = df_tv_long['tvcharttype'].str[:-3] # Remove the "-tv" suffix
df_tv_long['chart_id'] = df_tv_long['tvcharttype'] + df_tv_long['block'].apply(str) # Add the Block number to the end
df_tv_long.drop('block', axis=1, inplace=True)

df_long = pd.merge(df_long,df_tv_long, on='chart_id')
```

```python
df_long.columns
```

Now create the log mean error field

```python
def logMeanError(error):
    return np.log2(abs(error) + 1/8)

df_long['error'] = df_long['judgement'] - df_long['trueValue']
df_long['logMeanError'] = df_long['error'].apply(logMeanError)
```

Combine all the comments into one subset and export them as a .csv file for easy reading

```python
# subset of comments
comments = df.comments[pd.notnull(df['comments'])]
comments.to_csv('comments.csv')










#Delete this?
#pd.options.display.max_colwidth = 100
```

### 2.B. Comparison to Irish population

Read in census data and fit survey responses to the same format

```python
sex = pd.read_csv('censusData/sex.csv')
edu = pd.read_csv('censusData/education.csv')
age = pd.read_csv('censusData/age.csv')

samplesex = pd.DataFrame(df['gender'].value_counts()/len(df.gender[pd.notnull(df.gender)]))
samplesex.reset_index(inplace = True)
samplesex.drop([2], inplace=True)

sampleedu = pd.DataFrame(df['education'].value_counts()/len(df.education[pd.notnull(df.education)]))
sampleedu.reset_index(inplace = True)
sampleedu.drop([5], inplace=True)
```

It is a little more complicated to transform the survey age data becuase it must be sorted into the binned format given by the census data

```python
# Create the bins
bins = [5, 13, 19, 25, 35, 45, 55, 65, 75]

# Create a for loop to sum up the amount of participants who fell into each bin
binnedage = []
xold = 0
for x in bins:
    binnedage.append(len(df.age[df.age > xold][df.age < x]))
    xold=x
binnedage.append(0)
    
# Convert this list into a dataframe, label the bins
binnedage = pd.DataFrame(binnedage)
binnedage.columns = ['number']
binnedage['age'] = age['age']

# and then normalise it by dividing by the people who gave their age

binnedage['percent'] = binnedage.number/len(df.age[pd.notnull(df.age)])
```

```python
# Make the PairGrid
g = sns.PairGrid(tdf.sort_values("Total_Error", ascending=False),
                 x_vars=tdf.columns[-15:-9], y_vars=["timestamp"],
                 size=10, aspect=.25)

# Draw a dot plot using the stripplot function
g.map(sns.stripplot, size=10, orient="h",
      palette="Reds_r", edgecolor="gray")

# Use the same x axis limits on all columns and add better labels
g.set(xlim=(-10, 10), ylabel="")

# Use semantically meaningful titles for the columns
titles = ["Pie", "Donut","Treemap", "Bar", "Stacked", "Funnel"]

for ax, title in zip(g.axes.flat, titles):

    #Set a different title for each axes
    ax.set(title=title)

    # Make the grid horizontal instead of vertical
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)
    ax.set(xticklabels='')
    ax.set(yticklabels='')

sns.despine(left=True, bottom=True)

plt.savefig('../output/99Extra.png', dpi = 300)
```
